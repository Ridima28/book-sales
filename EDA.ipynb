```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Books_Data_Clean.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df[df["Publishing Year"]> 1900]
```

```python
df.describe()
```

```python
df = df.drop(columns='index', inplace = True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset='Book Name', inplace = True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of Publishing Year")
plt.savefig("Publishing_year_hist.png")
plt.show()
```

```python
df["genre"].value_counts().plot(kind = 'bar')
plt.xlabel("Genre")
plt.ylabel("Number of Books")
plt.title("Number of books in each genre ")
plt.savefig("Total_books_in_each_genre.png")
plt.show()
```

```python
df.groupby("Author")['Book_average_rating'].mean().sort_values(ascending = False).reset_index()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x = 'genre', y = "Book_ratings_count", data = df)
plt.xlabel("Genre")
plt.ylabel("Books Rating Count")
plt.title("Box plot of rating count for each genre")
plt.savefig("Box_plot_of_rating.png")
plt.show()
```

```python
plt.scatter(df["sale price"], df["units sold"])
plt.xlabel("Sales Price")
plt.ylabel("Units Sold")
plt.title("Scatter Plot of Sale Price vs Units sold")
plt.savefig("Scatter_plot.png")
```

```python
lang_counts = df["language_code"].value_counts().sort_values(ascending = True)
```

```python
plt.pie(lang_counts, labels= lang_counts.index, startangle=90, autopct = '%1.1f%%' )
plt.title("Language Distribution of books")
plt.savefig("Language distribution")
plt.show()
```

```python
df.groupby('Publisher ')["publisher revenue"].sum().sort_values(ascending = False).reset_index()
```

```python
df.groupby("Author_Rating")['Book_average_rating'].mean().reset_index()
```

```python
df.groupby("language_code").size().sort_values(ascending= False).reset_index()
```

```python
plt.scatter(df["Book_average_rating"], df["Book_ratings_count"])
plt.xlabel("Book Average Rating")
plt.ylabel("Book Rating count")
plt.title("Scatter plt of average rating vs rating count")
plt.savefig("average rating vs rating count.png")
plt.show()
```

```python
total_gross_scale = df.groupby("Author")["gross sales"].sum()
```

```python
total_gross_scale.sort_values(ascending=False).head(10).plot(kind= 'bar')
plt.xlabel("Author")
plt.ylabel("Sales")
plt.savefig("Bar graph of author and sales")
plt.show()
```
